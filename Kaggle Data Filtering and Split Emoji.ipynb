```python
import pandas as pd
import emoji 
```

```python
def emoji2name(e):
    e_name = emoji.demojize(e).replace(':','')
    e_name = e_name.replace('_',' ')
    
    
    skin_tone = [' light skin tone',' medium-light skin tone',' medium skin tone',' medium-dark skin tone'
             ,' dark skin tone']
    
    for st in skin_tone:
        if st in e_name:
            e_name = e_name.replace(st,'')
            break
    

    return e_name

selected_emoji = [
    'face with tears of joy', # funny happy
    
    'red heart', #love
    
    'broken heart', #no love
    
    'thumbs up', #encouragement
    
    'smiling face with smiling eyes', #happy
    
    'loudly crying face', #pure sad
    
    'clapping hands', #congradualation
    
    'fire', #hot /sexy
    
    'face screaming in fear', #shock
    
    'pile of poo', #non-sense, disapproval
    
    'face with symbols on mouth', #anger
    
    'eggplant', #horny
    
    'face savoring food', # craving
    
    'hundred points', #approval
    
    'folded hands', #pray
    
    
]
```

```python
f =  open('emojitweets-01-04-2018.txt', encoding='utf-8')

content_list = []


for line in f:
    emoji_list_in_text = [emoji2name(e) for e in emoji.distinct_emoji_list(line)]
    
                
    for e_name in emoji_list_in_text:
        if e_name in selected_emoji:
            content = {'emoji_name': e_name, 'cleaned content': line}
            content_list.append(content)
    
    
```

```python
kaggle_data = pd.DataFrame(content_list)
kaggle_data
# len(kaggle_data['emoji_name'].unique())
```
